```python
df = df.rename(columns={"Url_spotify" : "Spotify URL",
                        "Album_type" : "Album Type",
                        "Uri" : "URI",
                        "Duration_ms" : "Duration MS",
                        "Url_youtube" : "YouTube URL",
                        "official_video" : "Official Video"
                        
})
df.head(5)

```

```python
##Data conversion remove nan values
df['Licensed'] = df['Licensed'].replace({True: 1, False: 0})
df['Official Video'] = df['Official Video'].replace({True: 1, False: 0})
df = df.dropna(subset=['Licensed', 'Official Video'])
df['Licensed'] = df['Licensed'].astype(bool)
df['Official Video'] = df['Official Video'].astype(bool)
```

```python
df.isnull().sum()
df['Stream'] = df['Stream'].fillna(df.groupby(['Artist', 'Album'])['Stream'].transform('mean'))
df['Likes'] = df['Likes'].fillna(df.groupby('Artist')['Likes'].transform('mean'))
df['Comments'] = df['Comments'].fillna(df.groupby(['Artist', 'Album'])['Comments'].transform('mean'))
df['Description'] = df['Description'].fillna('No Description')
df = df.dropna()
```

# Graphs of Queries


```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,10))

# First chart: Total Views

views_info = df.groupby("Artist").agg({"Views": "sum"}).sort_values(by="Views", ascending=False).head(10)
ax1.bar(views_info.index, views_info["Views"], color=["#FF6347", "#4169E1", "#FFA500", "#20B2AA", "#9370DB"])
ax1.set_xticklabels(views_info.index, rotation=90)
ax1.set_xlabel("Artist")
ax1.set_ylabel("Toplam ƒ∞zlenme Sayƒ±sƒ±")
ax1.set_title("Sanat√ßƒ±lara G√∂re Toplam YouTube ƒ∞zlenme Sayƒ±sƒ±")
# Second chart: Total Number of Likes

likes_info = df.groupby("Artist").agg({"Likes": "sum"}).sort_values(by="Likes", ascending=False).head(10)
ax2.bar(likes_info.index, likes_info["Likes"], color=["#FF6347", "#4169E1", "#FFA500", "#20B2AA", "#9370DB"])
ax2.set_xticklabels(likes_info.index, rotation=90)
ax2.set_xlabel("Artist")
ax2.set_ylabel("Toplam Beƒüeni Sayƒ±sƒ±")
ax2.set_title("Sanat√ßƒ±lara G√∂re Toplam YouTube Beƒüeni Sayƒ±sƒ±")

# Third chart: Total Streams

stream_info = df.groupby("Artist").agg({"Stream": "sum"}).sort_values(by="Stream", ascending=False).head(10)
ax3.bar(stream_info.index, stream_info["Stream"], color=["#FF6347", "#4169E1", "#FFA500", "#20B2AA", "#9370DB"])
ax3.set_xticklabels(stream_info.index, rotation=90)
ax3.set_xlabel("Artist")
ax3.set_ylabel("Toplam Akƒ±≈ü Sayƒ±sƒ±")
ax3.set_title("Sanat√ßƒ±lara G√∂re Toplam Spotify Akƒ±≈ü Sayƒ±sƒ±")

plt.tight_layout()
plt.show()
```

# Review by Albums, Songs and Platforms

```python
album_type_counts = df["Album_type"].value_counts()

sns.set(style="white")


plt.figure(figsize=(7, 5))


plt.pie(
    album_type_counts,
    labels=album_type_counts.index,
    autopct='%1.1f%%',
    startangle=140,
)
plt.axis('equal')
plt.title('Alb√ºme g√∂re Daƒüƒ±lƒ±m')
plt.show()
```

```python
data = df.groupby("Artist").agg({"Stream" : "mean", 
                                 "Views" : "mean"}).sort_values(by = [
                           "Stream", "Views"], ascending = [False, False]).head(30).reset_index()
data
```

```python

data_melted = data.melt(id_vars='Artist', 
                        value_vars=['Stream', 'Views'], 
                        var_name='Type', 
                        value_name='Value')


plt.figure(figsize=(10, 10))

sns.barplot(y='Artist', 
            x='Value', 
            hue='Type', 
            data=data_melted,
            palette='viridis') 

plt.title('Sanat√ßƒ±lara G√∂re Ortalama Spotify Stream ve YouTube Views Sayƒ±larƒ±')
plt.xlabel('Ortalama Deƒüer')
plt.ylabel('Sanat√ßƒ±')

sns.despine()

plt.show()
```

```python
data2 = df.groupby("Title").agg({"Stream" : "mean", 
                                 "Views" : "mean"}).sort_values(by = [
                           "Stream", "Views"], ascending = [False, False]).head(20).reset_index()
data2
```

```python

data2_melted = data2.melt(id_vars='Title', 
                          value_vars=['Stream', 'Views'], 
                          var_name='Type', 
                          value_name='Value')


plt.figure(figsize=(10, 10))

sns.barplot(y='Title', 
            x='Value', 
            hue='Type', 
            data=data2_melted,
            palette='Set2')  

plt.title('≈ûarkƒ±lara G√∂re Ortalama Spotify Stream ve YouTube Views Sayƒ±larƒ±')
plt.xlabel('Ortalama Deƒüer')
plt.ylabel('≈ûarkƒ±')

sns.despine()

plt.show()
```
